```python
import pandas as pd
deliveries = pd.read_csv("deliveries.csv")
orders = pd.read_csv("orders.csv")
complaints = pd.read_csv("complaints.csv")
customers = pd.read_csv("customers.csv")
drivers = pd.read_csv("drivers.csv")
incidents = pd.read_csv("incidents.csv")
```

```python
deliveries[['manual_route_override_count','customer_rating_post_delivery']].corr()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = deliveries[['manual_route_override_count']]

y = deliveries['customer_rating_post_delivery']
```

```python
model = LinearRegression()

model.fit(X, y)
```

```python
regression_data = deliveries[['manual_route_override_count', 'customer_rating_post_delivery']].dropna()

X = regression_data[['manual_route_override_count']]
y = regression_data['customer_rating_post_delivery']

model = LinearRegression()
model.fit(X, y)
```

```python
model.coef_
```
